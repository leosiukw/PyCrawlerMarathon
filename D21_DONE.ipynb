## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

## 1. 取出今天所有的新聞

```python
# 打開瀏覽器

from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list-2020-6-26-0.htm")


# 每個兩秒鐘自動往下滑

import time
for i in range(60):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
news = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    news.update({d.find(class_="date").text : d.find_all('a')[-1].text})

# 時間轉換：https://stackoverflow.com/questions/10256093/how-to-convert-ctime-to-datetime-in-python
    
# 關閉瀏覽器
browser.quit();

print(news)
```

```python
import datetime

for key, value in news.items():
    d = datetime.datetime.strptime(key, "%Y/%m/%d %H:%M")
    if (d.year==2020) and (d.month==6) and (d.day==26):
        print(key, value)
```

## 2. 取出現在時間兩小時內的新聞

```python
for key, value in news.items():
    d = datetime.datetime.strptime(key, "%Y/%m/%d %H:%M")
    if (d.year==2020) and (d.month==6) and (d.day==26) and ((d.hour==23) or (d.hour==22)):
        print(key, value)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list-2020-6-23-0.htm")

# 每個兩秒鐘自動往下滑

import time
for i in range(60):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
news = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    news.update({d.find(class_="date").text : d.find_all('a')[-1].text})

# 時間轉換：https://stackoverflow.com/questions/10256093/how-to-convert-ctime-to-datetime-in-python
    
# 關閉瀏覽器
browser.quit();

for key, value in news.items():
    d = datetime.datetime.strptime(key, "%Y/%m/%d %H:%M")
    if (d.year==2020) and (d.month==6) and (d.day==23) and ((d.hour==15) or (d.hour==16)):
        print(key, value)
```
