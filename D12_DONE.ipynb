### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text)

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find("em").text, d.find('a').text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
date_list = []
category_list = []
title_list = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    date_list.append(d.find(class_="date").text)
    category_list.append(d.find("em").text)
    title_list.append(d.find('a').text)

import pandas as pd

data = {'date': date_list,
       'category': category_list,
       'title': title_list}

df = pd.DataFrame(data)

df
```

### ③ 哪一個類別的文章最多

```python
df['category'].value_counts()
```
